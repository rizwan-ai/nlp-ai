## Sentiment Analysis

Sentiment analysis is the automated process of tagging data according to their sentiment, such as positive, negative and neutral. Sentiment analysis allows companies to analyze data at scale, detect insights and automate processes.

### What is Sentiment Analysis?

Sentiment analysis is a natural language processing technique that identifies the polarity of a given text. There are different flavors of sentiment analysis, but one of the most widely used techniques labels data into positive, negative and neutral.

Sentiment analysis allows processing data at scale and in real-time. For example, do you want to analyze thousands of tweets, product reviews or support tickets? Instead of sorting through this data manually, you can use sentiment analysis to automatically understand how people are talking about a specific topic, get insights for data-driven decisions and automate business processes.

Sentiment analysis is used in a wide variety of applications, for example:

- Analyze social media mentions to understand how people are talking about your brand vs your competitors.
- Analyze feedback from surveys and product reviews to quickly get insights into what your customers like and dislike about your product.
- Analyze incoming support tickets in real-time to detect angry customers and act accordingly to prevent churn.

### EDA on a Twitter dataset

```python
# Load the Twitter dataset from NLTK
positive_tweets = twitter_samples.strings("positive_tweets.json")
negative_tweets = twitter_samples.strings("negative_tweets.json")
```

```python
print(f"Number of positive tweets: {len(positive_tweets)}")
print(f"Number of negative tweets: {len(negative_tweets)}")
```

```python
print(f"The type of all positive_tweets is: {type(positive_tweets)}")
print(f"The type of a tweet entry is: {type(negative_tweets[0])}")
```

```python
for tweet in positive_tweets[:5]:
    if '#' not in tweet:
        print(tweet)
```

```python
# Display a random negative tweet
negative_tweets[random.randint(0, 4999)]
```

```python
# Display 5 random positive tweets
for i in range(5):
    print(positive_tweets[random.randint(0, 4999)])
```

```python
# Data for plotting
categories = ["Positive", "Negative"]
counts = [len(positive_tweets), len(negative_tweets)]
```

```python
# Plotting the bar chart
plt.figure(figsize=(6, 4))
plt.bar(categories, counts, color=['red', 'orange'], width=0.5)
plt.title('Count of Positive and Negative Tweets')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting the horizontal bar chart
plt.figure(figsize=(6, 4))
plt.barh(categories, counts, color=['red', 'orange'])
plt.xlabel('Count')
plt.title('Count of Positive and Negative Tweets')
plt.show()
```

```python
# Plotting the pie chart
plt.figure(figsize=(6, 4))
plt.pie(counts, labels=categories, colors=['red', 'orange'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Positive and Negative Tweets')
plt.show()
```

```python
# Generate random data for positive and negative tweets (simulated time series)
np.random.seed(0)
num_days = 30
positive_counts = np.random.randint(50, 100, num_days)
negative_counts = np.random.randint(30, 80, num_days)

# Data for plotting
days = np.arange(1, num_days + 1)

# Plotting the line plot
plt.figure(figsize=(10, 6))
plt.plot(days, positive_counts, label='Positive Tweets', marker='o')
plt.plot(days, negative_counts, label='Negative Tweets', marker='s')
plt.xlabel('Day')
plt.ylabel('Count')
plt.title('Positive and Negative Tweets Over Time')
plt.xticks(np.arange(1, num_days + 1, 5))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
